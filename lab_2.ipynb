```python
import spacy
```

here we are installing and importing the spacy packages in python

```python
import requests

url = "https://raw.githubusercontent.com/bratzlernathan-cloud/lis4693/refs/heads/main/lab-1/transcript.txt"
response = requests.get(url)
response.raise_for_status()
text = response.text
```

importing requests so we can get the transcript from github  

```python
print("Number of characters:", len(text))

print(text[:100])
```

printing the total number of characters and printing the first 100 characters of the transcript

```python
nlp = spacy.blank("en")
nlp.add_pipe("sentencizer")

doc = nlp(text)

sentences = list(doc.sents)

print("Number of sentences:", len(sentences))

print("\nFirst 3 sentences:")
for sent in sentences[:3]:
  print(sent)
```

we are printing the first 3 sentences of the transcript. Unfortunately my transcript did not have punctuation so the whole transcript file is considered one large sentence.

```python
words = [token.text.lower() for token in doc if token.is_alpha]

print("Total words:", len(words))
print("Unique words:", len(set(words)))
```

printing the total number of words and the unique number of words

```python
from collections import Counter

word_freq = Counter(words)

print("Top 10 most frequent words:")
for word, count in word_freq.most_common(10):
  print(word,count)
```

this is printing the top 10 most frequently used words

```python
nlp2 = spacy.load("en_core_web_sm")

doc2 = nlp2(text)

print("Named Entities:")
for ent in doc2.ents:
  print(ent.text, ent.label_)
```

it correctly named three quarters as a number, jack and nelson as people, but it labeled cj as norp instead of another person.

```python
from spacy.matcher import PhraseMatcher

matcher = PhraseMatcher(nlp2.vocab, attr= "LOWER")

phrases = ["eggs", "nicely", "heat"]

patterns = [nlp2(p) for p in phrases]

matcher.add("eggs", patterns)

matches = matcher(doc2)

print("Matches found:")
for match_id, start, end in matches:
  print(doc2[start:end])
  print("Sentence:", doc2[start].sent)
```
